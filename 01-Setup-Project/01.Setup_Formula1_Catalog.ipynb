# Create Formula1 Catalog

```python
%sql
CREATE CATALOG IF NOT EXISTS formula1
MANAGED LOCATION 'abfss://formula1@bronzeazureadlsstorage.dfs.core.windows.net/landing/'
COMMENT 'This is the main catalog for the formula1 project';
```

# Create Schemas

```python
%sql
CREATE SCHEMA IF NOT EXISTS formula1.landing
MANAGED LOCATION 'abfss://formula1@bronzeazureadlsstorage.dfs.core.windows.net/landing'
COMMENT 'landing schema';

CREATE SCHEMA IF NOT EXISTS formula1.bronze
MANAGED LOCATION 'abfss://formula1@bronzeazureadlsstorage.dfs.core.windows.net/landing/bronze'
COMMENT 'bronze schema';

CREATE SCHEMA IF NOT EXISTS formula1.silver
MANAGED LOCATION 'abfss://formula1@bronzeazureadlsstorage.dfs.core.windows.net/landing/silver'
COMMENT 'silver schema';

CREATE SCHEMA IF NOT EXISTS formula1.gold
MANAGED LOCATION 'abfss://formula1@bronzeazureadlsstorage.dfs.core.windows.net/landing/gold'
COMMENT 'gold schema';
```
